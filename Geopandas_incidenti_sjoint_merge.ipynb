```python
import geopandas as gpd
import contextily as ctx
import matplotlib.pyplot as plt
```

```python
df=gpd.read_file("https://github.com/SimoneFinessi/Dati/blob/main/incidenti.zip?raw=true")
```

```python
df.crs
```

```python
ax=df.to_crs(3857).plot(figsize=(12,12))
ctx.add_basemap(ax=ax)
```

```python
comuni=gpd.read_file("/content/drive/MyDrive/istat2022/Com01012022_g.zip")
comuni
```

```python
ax=comuni[comuni.intersects(df.to_crs(32632).unary_union)].to_crs(3857).plot(edgecolor="black",alpha=0.3)
df.to_crs(3857).plot(ax=ax,color="Red",markersize=1)
ctx.add_basemap(ax=ax)
```

```python
joint=gpd.sjoin(df.to_crs(32632),comuni,predicate="intersects",how="left")
```

```python

n_fonta_quar=joint.groupby("COMUNE")[["COD_RIP"]].count().reset_index()#faccio il groupby di NIL per contare quante fontane ci sono per q
```

```python
finale=comuni.merge(n_fonta_quar,on="COMUNE")
```

```python
ax=finale[finale.intersects(df.to_crs(32632).unary_union)].to_crs(3857).plot(edgecolor="black",alpha=0.3,column="COD_RIP_y",legend=True)
df.to_crs(3857).plot(ax=ax,color="Red",markersize=1)
ctx.add_basemap(ax=ax)
```
